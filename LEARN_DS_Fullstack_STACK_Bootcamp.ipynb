## Modelagem dos Dados

image.png

Após a compreensão do modelo, deve-se começar a trabalhar na modelagem dos dados.

A **modelagem dos dados** é a etapa de entender os dados do projeto, quais as tabelas serão usadas, quais dados serão extraídos de quais fontes, de acordo com o problema apresentado. O objetivo é obter um *Result Set*, um conjunto de dados ótimos para atender o problema final.

Os dados podem ser provenientes de JSON, que possuem uma estrutura de dicionário. Em arquivos CSV e XLS, os dados estão estruturados de forma tabular, mas não necessariamente possuem um único tipo de dado em uma coluna. Ou seja, podem existir diversos ruídos em uma mesma coluna. Nos dados em SGBD, os dados estão estruturados de forma tabular, mas estão armazenados dentro de um sistema de Banco de Dados.

O Airflow auxilia na criação das queries dos diversos formados de dados, como do gereciador de dados, dos arquivos de planilha ou do arquivo JSON.

### **5. MySQL Server**

O Container funciona como uma micro-máquina no local.

Criação de uma nova instância do container MySQL, usando a imagem oficial do MySQL, e configurando a senha raiz para "stack", expondo a porta 3306 do container como 3307 no host e executando o container em background.

`docker run -name mysqlbd1 -e MYSQL_ROOT_PASSWORD=stack -p "3307:3306" -d mysql`


> * `-name mysqlbd1`: Atribui um nome personalizado ao container, neste caso "mysqlbd1".
> 
> * `-e MYSQL_ROOT_PASSWORD=stack`: Define a senha raiz para o MySQL dentro do container.
> 
> * `-p "3307:3306"`: Liga a porta 3307 do host a porta 3306 do container. Isso permite que o banco de dados seja acessado através da porta 3307 no host. A porta 3306 é a porta padrão do MySQL. O uso da porta 3307 no host é importante para não criar incompatibilidade com a porta do MySQL no Host.
> 
> * `-d`: rodará em background
> * `mysql`: Especifica a imagem a ser usada para o container, no caso, a imagem oficial do MySQL.

### **6. Data Lake**

O MinIO é um servidor de Block Storage, sendo um repositório de diversos discos, especializado em armazenar dados não-estruturados ou semi-estruturados.

> Ele é uma solução Open-Source e utiliza do protocolo da Amazon S3

Por termos dados em JSON, CSV, planilhas, o MinIO ajuda a gerenciar todos estes dados, sendo mais interessante do que usar um servidor de arquivos.

Um **servidor de arquivos** é um dispositivo ou software que é responsável por armazenar, gerenciar e compartilhar arquivos em uma rede. Alguns exemplos são: Google Drive, Amazaon S3, Microsoft One, entre outros. São conhecidos como armazenamento em nuvem.


1. Criação do diretório `datalake`
    * `mkdir datalake`
2. Instância do MinIO
    * No Powershell
    * `docker run -name minio -d -p 9000:9000 -p 9001:9001 -v "$PWD/datalake:/data" minio/minio server /data --console-address ":9001"`

    > * O parâmetro ` -v "$PwD/datalake:/data"` cria um volume compartilhado entre o host e o container, mapeando a pasta "datalake" no host para a pasta "/data" no container. Ou seja, tudo que for persistido no diretório /data no datalake, será salva na pasta /datalake do host.
    > * Este mapeamento de volume é importante para não perder dados ao ser desligado ou removido o Container. Assim, a persistência é local.
    > * O parâmetro `minio/minio` especifica a imagem do Minio a ser usada
    > * `server /data --console-address ":9001"` especifica o caminho para a pasta de dados.

### **7. Apache Airflow**

O Airflow será o orquestrador para automatizar o Data Pipeline. Como se fosse uma aplicação de ETL.

1. Criação do diretório `airflow`
2. Criação do diretório `airflow/dags`
    * Onde serão armazenados os arquivos de orquestração
3. Criação da instância do Airflow

```shell
mkdir airflow
cd airflow
mkdir airflow\dags
cd ..

docker run -d -p 8080:8080 -v "$PWD/airflow/dags:/opt/airflow/dags/" --entrypoint=/bin/bash --name airflow apache/airflow:2.1.1-python3.8 -c '(airflow db init && airflow users create --username admin --password stack --firstname nome_usuário lastname Lastname --role Admin --email admin@example.org); airflow webserver & airflow scheduler'
```

* `-d` executa em segundo plano

* `-p 8080:8080` mapeia as portas
* `-v "$PWD/airflow/dags:/opt/airflow/dags/`: mapeiamento de pastas. Isso permite que qualquer arquivo colocado nessa pasta no host seja acessível no container e vice-versa.
* `apache/airflow:2.1.1-python3.8`: cria uma instância da imagem Docker
* `--name airflow`: nome do container
* `--entrypoint=/bin/bash`: especifica qual o ponto de entrada do container deve ser o interpretador de comandos bash.
* `-c`: executa um comando Bash, que irá iniciar um banco de dados (db) do Apache Airflow, criará um usuário (admin) e senha (stack), além de informações sobre ele. Inicia também um Webserver e um Scheduler.

Em resumo, este comando está iniciando um novo container do Airflow e mapeando suas pastas de arquivos de fluxo de trabalho e configurações para o host, e iniciando o banco de dados, o webserver e o scheduler do airflow.
